```python
import pandas as pd
import ast
import re
```

```python
new_reviews = pd.read_csv('new_review.csv')
```

```python
review_df = pd.read_csv('australian_user_reviews.csv')
```

```python
review_df
```

```python
# Lista para almacenar los datos extraídos
new_data = []

# Iterar a través de las filas del DataFrame original
for index, row in review_df.iterrows():
    user_id = row['user_id']
    reviews_str = row['reviews']

    # Convertir la cadena de reseñas en una lista de diccionarios
    reviews_list = ast.literal_eval(reviews_str)

    # Iterar a través de las reseñas del usuario
    for review in reviews_list:
        posted = review['posted']
        

        # Agregar los datos extraídos a la lista
        new_data.append({'user_id': user_id, 'posted': posted})

# Crear un nuevo DataFrame con los datos extraídos
new_review_df = pd.DataFrame(new_data)
```

```python
new_review_df
```

```python
new_review_df.drop(columns='user_id', inplace=True)
```

```python
new_review_df
```

```python
new_review_df['posted'] = new_review_df['posted'].str.replace('.', '')
```

```python
# Mapear los nombres de los meses a números de mes
month_dict = {
    'January': '01',
    'February': '02',
    'March': '03',
    'April': '04',
    'May': '05',
    'June': '06',
    'July': '07',
    'August': '08',
    'September': '09',
    'October': '10',
    'November': '11',
    'December': '12'
}

# Función para transformar la columna 'posted' a formato YYYY-MM-DD
def transform_date(row):
    try:
        parts = row.split()
        month = month_dict[parts[1]]
        day = parts[2].rstrip(',')
        year = parts[-1] if parts[-1].isdigit() else '2023'  # Año de referencia para fechas incompletas
        
        # Verifica si el año es menor o igual a 31 (posiblemente el día)
        if int(year) <= 31:
            # Si es menor o igual a 31, asumimos que se trata del año y el día está faltante
            day, year = year, '2023'
        
        return f"{year}-{month}-{day.zfill(2)}"
    except:
        return None  # Si no se puede transformar, devuelve None o un valor de fecha nulo

# Aplicar la función y crear una nueva columna 'transformed_date'
new_review_df['transformed_date'] = new_review_df['posted'].apply(transform_date)
```

```python
new_review_df
```

```python
new_review_df.drop(columns='posted', inplace=True)
```

```python
new_review_df
```

```python
new_review_df['posted'] = new_review_df['transformed_date']
```

```python
new_review_df.drop(columns='transformed_date', inplace=True)
```

```python
new_reviews = pd.concat([new_reviews,new_review_df], axis=1)
```

```python
new_reviews.dropna(subset=['user_id'], inplace=True)
```

```python
new_reviews
```

```python
new_reviews.to_csv('new_review.csv',index=False)
```
